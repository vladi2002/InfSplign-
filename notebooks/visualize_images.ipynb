```python
benchmark = "visor"
folder = "sd1.4_relu"#sd1.4_3_attn_maps_masked_mean_smoothing"
images_folder = os.path.join("images", benchmark, folder)
obj_det_folder = os.path.join("objdet_results", benchmark, folder)
```

```python
import matplotlib.pyplot as plt
import os
from PIL import Image
import numpy as np

def plot_images_for_prompt(prompt, objdet_ann, images_folder, num_images_per_prompt=4, figsize=(20, 5)):
    fig, axes = plt.subplots(1, num_images_per_prompt, figsize=figsize)

    image_files = []
    objdet_list = []
    for i in range(num_images_per_prompt):
        filename = os.path.join(images_folder, f"{prompt}_{i}.png")
        image_files.append(filename)

        key = f"{prompt}_{i}"
        obj_det_info = objdet_ann.get(key)
        objdet_list.append(obj_det_info)

    for i, img_file in enumerate(image_files):
        if i >= num_images_per_prompt:
            break

        img = Image.open(img_file)
        objdet = objdet_list[i]
        det_classes = list(set(objdet["classes"]))
        sra = objdet["sra"]
        visor_spatial = "yes" if sra else "no"
        # print(objdet)
        # print("Detected classes", list(set(objdet["classes"])))
        ax = axes[i] if num_images_per_prompt > 1 else axes
        ax.imshow(np.array(img))
        ax.set_title(f"Image {i+1}, detected: {det_classes}, visor: {visor_spatial}")
        ax.axis('off')

    # Set the main title with the prompt
    plt.suptitle(prompt, fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.show()

# Example usage
def plot_multiple_prompts(prompts_list, objdet_ann, images_folder, num_images=4):
    for prompt in prompts_list:
        print(f"Prompt: {prompt}")
        plot_images_for_prompt(prompt, objdet_ann, images_folder, num_images)
        print("\n" + "-"*80 + "\n")
```
